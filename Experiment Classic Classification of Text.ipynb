```python
from sklearn.metrics import classification_report, confusion_matrix,\
    f1_score
def eval_results(y_true, y_pred):
    print(classification_report(y_true, y_pred,\
                          target_names=label_encoder.classes_))
    print(confusion_matrix(y_true, y_pred))
    print(f1_score(y_true, y_pred, average='weighted'))
```

```python
from collections import Counter
def reinfer_ranking(d2v):
    ranks = []
    for i in range(len(documents[:100])):
        inferred_vector = d2v.infer_vector(documents[i].words, steps=20)
        sims = d2v.docvecs.most_similar([inferred_vector],
                                        topn=10)
        try:
            rank = [docid for docid, sim in sims].index(i)
        except ValueError:
            rank = None
        ranks.append(rank)
    counts = Counter(ranks)
    counts.pop(None)
    return sorted(counts.items())
```

```python
reinfer_ranking(doc2vec_1)
```

#### Nearest neighbors of samples

```python
def similar_docs(d2v, words):
    inferred = d2v.infer_vector(words, steps=20)
    sims = d2v.docvecs.most_similar([inferred], topn=3)
    return list(map(lambda x: x[0], sims)), list(map(lambda x: x[1], sims))
```
