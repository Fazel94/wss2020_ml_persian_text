```python
from sklearn.preprocessing import LabelEncoder
import joblib


label_encoder = joblib.load('/home/fazel/env/thesis/label_encoder.joblib')
df = pd.read_json('/home/fazel/env/thesis/hamshahri_df.json')
df['cat_english'] = label_encoder.transform(df['cat_english'])
```

```python
from sklearn.model_selection import train_test_split

tokens_train, tokens_test,\
    label_train, label_test = train_test_split(df['text'], 
                                              df['cat_english'],
                                              random_state=43)
```

```python
del df
```

### Feature engineering 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer?
```

```python
tfidf = TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(lowercase=False,
                        tokenizer=lambda x: x,
                        smooth_idf=False,
                        max_df=10000,
                        min_df=50,
                       )
```

```python
X_train = tfidf.fit_transform(tqdm(tokens_train))
```

```python
X_train
```

```python
X_test = tfidf.transform(tqdm(tokens_test))
```

```python
X_test
```

### Learning

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=1000, verbose=True)
```

```python
lr.fit(X_train, label_train)
```

```python
lr.score(X_train, label_train)
```

```python
lr.score(X_test, label_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,\
    f1_score
def eval_results(y_true, y_pred):
    print(classification_report(y_true, y_pred,\
                          target_names=label_encoder.classes_))
    print(confusion_matrix(y_true, y_pred))
    print(f1_score(y_true, y_pred, average='weighted'))
```

```python
eval_results(label_test, lr.predict(X_test))
```
